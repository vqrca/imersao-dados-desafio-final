## **Projeto desenvolvido:**
**Drug discovery - Análise de dados e predição de mecanismos de ação**

O monitoramento da expressão gênica por meio de abordagens de transcriptômica (perfil de RNA) tornou-se uma ferramenta extremamente relevante em esforços para compreender processos biológicos complexos. A análise de resultados provenientes deste tipo de abordagem permite a seleção de potenciais alvos de medicamentos e o perfil da expressão gênica provavelmente desempenhará um papel cada vez mais importante na descoberta de medicamentos.

A descoberta de medicamentos evoluiu da identificação de substâncias ativas em medicamentos tradicionais para a busca direta por novos medicamentos utilizando os alvos moleculares específicos e seus mecanismos de ação *(Mechanisms of Action - MoA)*. Portanto, esse projeto permitirá a predição de mecanismos de ação das drogas contidas no dataset que combina a expressão gênica e os dados de viabilidade celular.

- **Expressão gênica:**

  Um genoma é o conjunto completo de informações genéticas de um organismo e fornece todas as informações de que o organismo necessita para funcionar. Nos organismos vivos, o genoma é armazenado em longas moléculas de DNA chamadas cromossomos. Pequenas seções de DNA, chamadas de genes, codificam o RNA e as moléculas de proteína exigidas pelo organismo.   

  Portanto, os genes codificam proteínas e as proteínas ditam a função celular. Então, os milhares de genes expressos em uma determinada célula determinam o que essa célula pode fazer. Além disso, cada etapa no fluxo de informações do DNA para o RNA para a proteína fornece à célula um ponto de controle potencial para autorregular suas funções, ajustando a quantidade e o tipo de proteínas que fabrica.

  A qualquer momento, a quantidade de uma proteína específica em uma célula reflete o equilíbrio entre as vias bioquímicas sintéticas e degradativas dessa proteína. Assim, o controle desses processos desempenha um papel crítico na determinação de quais proteínas estão presentes em uma célula e em quais quantidades. 

  Nesse sentido, a avaliação da expressão gênica para avaliação da ação de compostos é essencial. Podemos avaliar como um composto afeta a expressão de diversos genes, ou até mesmo como afeta a expressão global (todo o genoma) de um organismo. No caso deste projeto, vemos a análise do perfil de expressão gênica para investigar o efeito de moléculas na resposta celular.

  O perfil de expressão gênica mede os níveis de mRNA, mostrando o padrão de genes expressos por uma célula no nível de transcrição. Isso geralmente significa medir as quantidades relativas de mRNA em duas ou mais condições experimentais e, em seguida, avaliar quais condições resultaram na expressão de genes específicos. 

  Um perfil de expressão gênica nos diz como uma célula está funcionando em um momento específico. Isso ocorre porque a expressão do gene celular é influenciada por estímulos externos e internos, incluindo se a célula está se dividindo, quais fatores estão presentes no ambiente da célula e/ou os sinais que ela está recebendo de outras células.

   - *valores de expressão gênica positivos:* indicam que houve superexpressão dos genes

   - *valores de expressão gênica negativos:* indicam que houve subregulação dos genes

   - *valores de expressão gênica próximos de zero:* indicam que não houve modulação na expressão gênica, ou seja, o gene são foi nem superexpresso nem subregulado.


- **Viabilidade celular:**

  A viabilidade celular é uma medida da proporção de células vivas e saudáveis em uma população. Os ensaios de viabilidade celular são usados para determinar a saúde geral das células, otimizar a cultura ou as condições experimentais e para medir a sobrevivência celular após o tratamento com compostos, como durante uma triagem de drogas.
 
  Para calcular a viabilidade celular é feito o seguinte cálculo: 

  *número de células viáveis/número total de células*


- **Mecanismos de Ação (MoA):**

  Termo usado para descrever como um medicamento ou outra substância produz um efeito no corpo. Por exemplo, o mecanismo de ação de uma droga pode ser como ela afeta um alvo específico em uma célula, como uma enzima, ou uma função celular, como o crescimento celular. Conhecer o mecanismo de ação de um medicamento pode ajudar a fornecer informações sobre a segurança do medicamento e como ele afeta o corpo. Também pode ajudar a identificar a dose certa de um medicamento e quais pacientes têm maior probabilidade de responder ao tratamento.

###**Distribuição dos genes e células**
É possível analisar a distribuição de diversos genes (colunas g0 a g771) e células (colunas c0 a c99) de forma randômica:

```python
#Definir quem são os genes e quem são as células:
genes = [g for g in dados.columns if g.startswith("g")]
print(f"Número de genes: {len(genes)}")
cells = [c for c in dados.columns if c.startswith("c")]
print(f"Número de células: {len(cells)}")
```

A análise randômica mostra que os genes apresentam, no geral, uma distribuição normal, com valores de expressão variando entre valores negativos e positivos, indicando que houve modulação na expressão gênica.

```python
#plotando a matriz de correlação dos genes

corr = dados.loc[:, 'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap=sns.color_palette("Spectral", as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, 
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

No geral genes não estão correlacionados, pois cada um tem uma função específica e seus valores de expressão dependem da ação das drogas.
O mesmo foi feito para correlacionar os tipos celulares:

```python
#plotando a matriz de correlação das células
corr_celular = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_celular, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_celular, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

O valores de viabilidade parecem ser muito próximos, pois a correlação dos tipos celulares é alta.

## **Aplicação de modelos de Machine Learning**

### **Problema para ser resolvido:** Nós temos 100 tipos de células que foram tratadas com diferentes compostos. Os Dados de expressão gênica e viabilidade celular foram coletados e analisados para entender a atividade biológica desses compostos. Portanto, o objetivo final deste projeto é prever o MoA (Mecanismo de ação) de novos compostos com base na expressão gênica e nas informações de viabilidade celular.

###**Testes realizados:**
- LogisticRegression
- DummyClassifier
- DecisionTreeClassifier
- RandomForestClassifier
- GradientBoostingClassifier
- HistGradientBoostingClassifier

Definindo os dados que serão treinados:

```python
dados_combinados.head()
```

```python
dados_resultados.head()
```

```python
#definição dos dados que serão treinados: x e y. 
#A proporção entre treino e teste, varia de acordo com o volume de dados que temos. 
#A proporção foi definida como 0,2 (20%) para teste e 80% para treino.

from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)
```

###**LogisticRegression**


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_rlogistica = LogisticRegression(max_iter=1000) 
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

**O modelo está acertando ~62% das previsões. Este valor é baixo para predizer os MoAs!**

###**DummyClassifier**

```python
#o modelo Dummy vai chutar que todos os eventos da base de dados assumem aquele valor no target. Assim, temos:
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

```python
#Para entender a acurácia do DummyClassifier é necessário um um value_counts normalizado:
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

**O valor mais frequente (estratégia usada no DummyClassifier) representa 60% da base de dados. Isso quer dizer que, caso o modelo chutasse a mesma classe para todos os eventos da base de teste, teríamos um acerto de 60% dos casos.**

###**DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier


x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3) #max_depth: quantidade de camadas (profundidade da árvore)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
#Plotando a imagem da árvore gerada:
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

A figura é dividida em 4 níveis:

- primeiro nível: é composto por apenas um quadro que é chamado de raiz. Os melhores atributos que podem dividir a nossa base de dados em dois conjuntos distintos

- Entre o primeiro e o segundo nível (e nos demais níveis), temos duas setas: uma para a direita e outra para a esquerda. Essas setas são chamadas de ramos e subdividem o nível acima em dois outros conjuntos que separam novamente os dados em mecanismo de ação ativado ou não ativado, a partir de uma decisão.

- O último nível da árvore: apresenta quadros que mostram características que devem ser levadas em conta na hora da classificação e se dividem em class = ativado e class = não ativado.

**Este modelo teve uma acurácia de 61%**

###**RandomForestClassifier**
A função que treina um modelo de Random Forest (RandomForestClassifier()) não aceita que tenhamos variáveis em formato de string. Por isso, precisamos tratar as variáveis que são strings e transformá-las em números (ou no formato float64)

```python
#Para transformar os dados = duas novas colunas binárias:
pd.get_dummies(dados_combinados['tratamento'])
```

```python
#é possível transformar todas as nossas variáveis de interesse: tratamento, dose e tempo.
pd.get_dummies(dados_combinados, columns=['tratamento', 'dose', 'tempo'])
```

```python
#Para retirar essas variáveis da nossa base de dados, usamos a função drop, declaramos as variáveis a serem retiradas 
#em forma de uma lista e identificamos que são colunas através do parâmetro axis = 1.
dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier


x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_randomforest = RandomForestClassifier()
modelo_randomforest.fit(x_treino, y_treino)
modelo_randomforest.score(x_teste, y_teste)
```

**A acurácia obtida com esse modelo foi de ~68,0%** e, ela é, aproximadamente, 7% melhor que a acurácia que conseguimos com o DummyClassifier.







###**GradientBoostingClassifier**

```python
from sklearn.datasets import make_hastie_10_2
from sklearn.ensemble import GradientBoostingClassifier

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y = dados_combinados['ativo_moa'] 

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=42)

clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
clf.fit(x_treino, y_treino)
clf.score(x_teste, y_teste)
```

**O resultado também não melhorou com o GradientBoostingClassifier: acurácia de 67.2%**

###**HistGradientBoostingClassifier**

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.datasets import make_hastie_10_2

x = dados_combinados.drop(['id', 'n_moa', 'ativo_moa', 'composto'], axis=1)
x = pd.get_dummies(x, columns=['tratamento', 'dose', 'tempo'])
y = dados_combinados['ativo_moa'] 

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=42)

clf = HistGradientBoostingClassifier(max_iter=100).fit(x_treino, y_treino)
clf.score(x_teste, y_teste)
```

**Obtivemos uma melhora para 69% de acurácia.**


Mesmo chegando próximo de 70% essa acurácia ainda náo é a ideal e suficiente para ser aplicada na vida real.
